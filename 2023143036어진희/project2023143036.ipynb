```python
import pandas as pd
import matplotlib.pyplot as plt

# Data input
data = {
    "연도": [2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009,
           2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023],
    "대한민국": [None, 4.7, 7.7, 3.1, 5.2, 4.4, 5.2, 5.8, 3.0, 0.8, 
                7.0, 3.7, 2.5, 3.3, 3.2, 2.9, 3.2, 3.4, 3.2, 2.3, -0.7, 4.6, 2.7, 1.4]
}

# Create DataFrame
df = pd.DataFrame(data)

# Drop missing data
df = df.dropna()

# Plotting the bar chart
plt.figure(figsize=(12, 6))
plt.bar(df["연도"], df["대한민국"], color='skyblue', label='대한민국 경제성장률')

plt.title('대한민국 연도별 경제성장률 변화', fontsize=16)
plt.xlabel('연도', fontsize=12)
plt.ylabel('경제성장률 (%)', fontsize=12)
plt.xticks(rotation=45)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


aa = pd.read_excel('경제성장률_불변가격__20241213060110.xlsx')
print(aa.columns)

try:
    years = aa['국가별']  # 국가명을 나타내는 열 (수정 필요 시 열 이름 확인 후 변경)
    growth_rate = pd.to_numeric(aa['2000'], errors='coerce')  # '2000' 열의 값 (숫자 변환)
except KeyError as e:
    print(f"열 이름을 확인하세요. 지정된 열이 없습니다: {e}")
    raise


valid_indices = ~growth_rate.isna()
years = years[valid_indices]
growth_rate = growth_rate[valid_indices]


plt.figure(figsize=(10, 6))
plt.bar(years, growth_rate, color='skyblue', label='경제성장률')

plt.title('연도별 경제성장률 변화', fontsize=16)
plt.xlabel('국가', fontsize=12)
plt.ylabel('경제성장률 (%)', fontsize=12)
plt.xticks(rotation=45)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
